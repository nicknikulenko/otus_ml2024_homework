# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib as plt
import seaborn
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips_data = seaborn.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
tips_data.head()
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(f'{tips_data.shape[0]=}')
```

### Проверьте есть ли пропуски в данных

```python
print(f'{tips_data.isnull().sum()=}')
```

### Посмотрите на распределение числовых признаков

```python
print(f'{tips_data.describe()=}')
```

### Найдите максимальное значение 'total_bill'

```python
print(f'{tips_data['total_bill'].max()=}')
```

### Найдите количество курящих людей

```python
print(f'{tips_data[tips_data['smoker'] == 'Yes']['smoker'].count()=}')
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print(f'{tips_data.groupby(['day'])['total_bill'].mean()=}')
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
print(f'{tips_data[tips_data['total_bill'] > tips_data['total_bill'].median()].groupby(['sex'])['tip'].mean()=}')
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips_data['smoker'] = tips_data['smoker'].map({'Yes': 1, 'No': 0})
print(f'{tips_data=}')
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.pyplot.hist(tips_data['total_bill'], color = 'red', edgecolor = 'black')
plt.pyplot.title('Histogram of total_bills')
plt.pyplot.xlabel('bill')
plt.pyplot.ylabel('freq')
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.pyplot.scatter(tips_data['total_bill'], tips_data['tip'])
plt.pyplot.title('Scatter of total_bills and tips')
plt.pyplot.xlabel('bill')
plt.pyplot.ylabel('tip')
```

### Постройте pairplot

```python
seaborn.pairplot(tips_data)
plt.pyplot.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
seaborn.pairplot(tips_data[['total_bill','tip']])
plt.pyplot.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.pyplot.hist(tips_data[tips_data['time'] == "Dinner"]['tip'], color = 'red', edgecolor = 'black')
plt.pyplot.hist(tips_data[tips_data['time'] == "Lunch"]['tip'], color = 'green', edgecolor = 'black')
plt.pyplot.title('Histogram of tip')
plt.pyplot.xlabel('tip')
plt.pyplot.ylabel('freq')
```

```python
dist1 = tips_data[tips_data['time'] == "Dinner"]['tip']
dist2 = tips_data[tips_data['time'] == "Lunch"]['tip']

fig, axs = plt.pyplot.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(dist1, color = 'red', edgecolor = 'black')
axs[1].hist(dist2, color = 'green', edgecolor = 'black')
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
dist1 = tips_data[tips_data['sex'] == "Male"]
dist2 = tips_data[tips_data['sex'] == "Female"]

fig, axs = plt.pyplot.subplots(1, 2, sharey=True, tight_layout=True)

sc1 = axs[0].scatter(dist1['total_bill'], dist1['tip'], c=dist1['smoker'], cmap = plt.colors.ListedColormap(['blue', 'red']))
leg1 = axs[0].legend(*sc1.legend_elements(), loc = 'upper left', title = 'smoker')
axs[0].add_artist(leg1)
axs[0].set_title('Male')
sc2 = axs[1].scatter(dist2['total_bill'], dist2['tip'], c=dist2['smoker'], cmap = plt.colors.ListedColormap(['green', 'pink']))
leg2 = axs[1].legend(*sc2.legend_elements(), loc = 'upper left', title = 'smoker')
axs[1].add_artist(leg2)
axs[1].set_title('Female')
```
